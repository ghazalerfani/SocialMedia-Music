```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_rows',300)
pd.set_option('display.max_columns',200)
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
df = pd.read_pickle('sony_music_tsne_gmm_dfCluster.pkl')
```

```python
df_noncluster_columns = ['Facebook', 'Instagram', 'Twitter', 'Youtube', 'age', 'artist_name_x',
       'created_12am-6am', 'created_12pm-6pm', 'created_6am-12pm',
       'created_6pm-12am', 'description_length', 'hashtag_count',
       'last_30_days', 'last_7_days', 'time_since_last_post', 'type_link',
       'type_photo', 'type_text', 'type_video', 'years_active',
       'num_days_to_track_release', 'genre_electronic', 'genre_hip_hop',
       'genre_pop', 'genre_r_and_b', 'genre_rock', 'genre_world_music']
```

```python
df.drop_duplicates(subset = df_noncluster_columns, inplace = True)
```

```python
df_og = pd.read_pickle('Combined_Social_Data_Pickle.pkl')
```

```python
#Only use data from 1/1/16 to 6/1/19
df_og = df_og[(df_og['created']>pd.to_datetime('01-01-2016')) & (df_og['created']<pd.to_datetime('06-01-2019'))]
```

```python
df_og['time_since_last_post'] = pd.to_timedelta(df_og['time_since_last_post'])
df_og['time_since_last_post'] = [x.total_seconds() for x in df_og['time_since_last_post']]
```

```python
df_og['crowdtangle_score'] = df_og['crowdtangle_score'].str.replace(',', '').astype(float)
```

```python
df_og = df_og[df_og['social_engagement_score'].notnull()]
```

```python
# merge with original dataset
df_merged = df.merge(df_og, how = 'left', on=df_noncluster_columns)
```

```python
df.shape
```

```python
df = df_merged[['Facebook', 'Instagram', 'Twitter', 'Youtube', 'age', 'artist_name_y',
       'created_12am-6am', 'created_12pm-6pm', 'created_6am-12pm',
       'created_6pm-12am', 'description_length', 'hashtag_count',
       'last_30_days', 'last_7_days', 'time_since_last_post', 'type_link',
       'type_photo', 'type_text', 'type_video', 'years_active',
       'num_days_to_track_release', 'genre_electronic', 'genre_hip_hop',
       'genre_pop', 'genre_r_and_b', 'genre_rock', 'genre_world_music', 
        'cluster_0', 'cluster_1', 'cluster_2', 'cluster_3', 'cluster_4','cluster_5', 
        'social_engagement_score', 'crowdtangle_score']]
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.shape
```

```python
cluster_columns = [col for col in df if col.startswith('cluster_')]
```

```python
# df_testing = df.groupby(['artist_name_x'], as_index=False)[cluster_columns].sum()
df['artist_name_y'].nunique()
```

```python
# get distribution of artist by cluster (for each artist = 1)
df_artists_by_cluster = df.groupby(['artist_name_y'], as_index=False)[cluster_columns].sum()
df_artists_by_cluster['clusters_sum'] = df_artists_by_cluster.sum(axis=1)
for c in cluster_columns:
    df_artists_by_cluster[c] = df_artists_by_cluster[c]/df_artists_by_cluster['clusters_sum']
df_artists_by_cluster = df_artists_by_cluster.drop(['clusters_sum'], axis=1)
df_artists_by_cluster.head()
```

```python
df_artists_by_cluster.columns.values.tolist()
```

```python
# merge all columns together for aggregate artist list
df_artists_all_other_columns = df.groupby(df['artist_name_y'], as_index=False).agg({"crowdtangle_score": "mean", "social_engagement_score": "mean", "Facebook": "sum", "Instagram": "sum", "Twitter": "sum", "Youtube": "sum", "age":"mean", "created_12am-6am": "sum", "created_12pm-6pm": "sum", "created_6am-12pm": "sum", "created_6pm-12am": "sum", "description_length": "mean", "last_7_days": "mean", "last_30_days": "mean", "time_since_last_post": "mean", "type_link": "sum", "type_photo": "sum", "type_text": "sum", "type_video": "sum", "years_active": "mean", "num_days_to_track_release": "mean", "genre_electronic": "sum", "genre_hip_hop": "sum", "genre_pop": "sum", "genre_r_and_b": "sum", "genre_rock": "sum", "genre_world_music": "sum"})
df_aggregate = df_artists_by_cluster.merge(df_artists_all_other_columns, on='artist_name_y')
```

```python
df_aggregate.to_csv('artists_final_aggregate.csv')
```

```python
dummies_platform = ['Facebook','Instagram','Twitter','Youtube']
x = df[dummies_platform].stack()
df.reset_index(inplace=True,drop=True) #necessary since the index numbers weren't starting at zero.
df['platform'] = pd.Series(pd.Categorical(x[x!=0].index.get_level_values(1)))
```

```python
# Aggregates per platform
df_aggregate_platform = df.groupby(df['platform'], as_index=False).agg({"crowdtangle_score": "mean", "social_engagement_score": "mean", "Facebook": "sum", "Instagram": "sum", "Twitter": "sum", "Youtube": "sum", "age":"mean", "created_12am-6am": "sum", "created_12pm-6pm": "sum", "created_6am-12pm": "sum", "created_6pm-12am": "sum", "description_length": "mean", "last_7_days": "mean", "last_30_days": "mean", "time_since_last_post": "mean", "type_link": "sum", "type_photo": "sum", "type_text": "sum", "type_video": "sum", "years_active": "mean", "num_days_to_track_release": "mean", "genre_electronic": "sum", "genre_hip_hop": "sum", "genre_pop": "sum", "genre_r_and_b": "sum", "genre_rock": "sum", "genre_world_music": "sum"})
df_aggregate_platform
```

```python
# Aggregate across all
df_aggregate_all = pd.DataFrame(df.agg({"crowdtangle_score": "mean", "social_engagement_score": "mean", "Facebook": "sum", "Instagram": "sum", "Twitter": "sum", "Youtube": "sum", "age":"mean", "created_12am-6am": "sum", "created_12pm-6pm": "sum", "created_6am-12pm": "sum", "created_6pm-12am": "sum", "description_length": "mean", "last_7_days": "mean", "last_30_days": "mean", "time_since_last_post": "mean", "type_link": "sum", "type_photo": "sum", "type_text": "sum", "type_video": "sum", "years_active": "mean", "num_days_to_track_release": "mean", "genre_electronic": "sum", "genre_hip_hop": "sum", "genre_pop": "sum", "genre_r_and_b": "sum", "genre_rock": "sum", "genre_world_music": "sum"}))
df_aggregate_all.T
```

```python
# Aggregate per cluster
df_clusters_summary = pd.DataFrame()
for c in cluster_columns:
    df_cluster = df_aggregate[df_aggregate[c]>0.28]
    df_cluster_agg = pd.DataFrame(df_cluster.agg({"artist_name_y": "count","crowdtangle_score": "mean", "social_engagement_score": "mean", "Facebook": "sum", "Instagram": "sum", "Twitter": "sum", "Youtube": "sum", "age":"mean", "created_12am-6am": "sum", "created_12pm-6pm": "sum", "created_6am-12pm": "sum", "created_6pm-12am": "sum", "description_length": "mean", "last_7_days": "mean", "last_30_days": "mean", "time_since_last_post": "mean", "type_link": "sum", "type_photo": "sum", "type_text": "sum", "type_video": "sum", "years_active": "mean", "num_days_to_track_release": "mean", "genre_electronic": "sum", "genre_hip_hop": "sum", "genre_pop": "sum", "genre_r_and_b": "sum", "genre_rock": "sum", "genre_world_music": "sum"})).T
    df_clusters_summary = pd.concat([df_clusters_summary, df_cluster_agg])
df_clusters_summary.reset_index(drop=True)
```

## Cluster Descriptions

### 0:
    -very low social engagement score (SES)
    -low Instagram posts
    -highest age and years active
    -highest post description length
    -mostly rock
### 1:
    -second highest SES
    -lowest years active (same as cluster 3
    -highest time since last post
    -mostly pop, electronic
### 2:
    -lowest # of posts 
    -more photos than other types of posts (vew few text posts)
    -second highest years active (and almost age) (after cluster 0)
### 3:
    -lowest age (and years active--same as cluster 3)
    -few FB posts
    -lowest number of days to track release
    -mostly R&B, HipHop
### 4:
    -highest SES
    -highest Twitter posts,
    -mostly Rock, R&B
### 5:
    -highest Instagram posts
    -very high Twitter posts
    -mostly HipHop
